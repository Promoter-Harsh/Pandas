```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [5,1,7,8,2]
plt.plot(x,y)
```

```python
plt.plot(x,y, color = "green", marker = "o")
plt.title("line plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x = [1,3,5,7,9]
line1 = [3,4,6,5,9,]
line2 = [2,5,3,7,9]
```

```python
plt.plot(x,line1, color ="green", marker = "o", linestyle = "--", label = 'line1')
plt.plot(x,line2, color ="green", marker = "o", linestyle = "--", label = 'line2 ')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line plot")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(x,line1, color ="green", marker = "o", linestyle = "--", label = 'line1')
plt.plot(x,line2, color ="green", marker = "o", linestyle = "--", label = 'line2 ')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line plot")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

```python
data1 = np.random.normal(0,1,1000)
data2 = np.random.normal(1,1,1000)
data3 = np.random.normal(2,1,1000)
```

```python
plt.hist([data1, data2, data3],bins = 10, color = ["red", "b", "g"], label = ["data1", "data2", "data3"])
plt.legend()
plt.xlabel("value")
plt.ylabel("frequency")
plt.title("histogram")
plt.show()
```

```python
langs = ["c", "python", "java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = "%1.1f%%")
plt.title("pie chart")
plt.legend( loc = "best", bbox_to_anchor = (1, 1))
plt.show()
```
